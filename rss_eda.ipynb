```python
import pandas as pd
```

создадим датасеты под (виды новостей и комании)

```python
nums = pd.read_csv('macro_joined_daily.csv', parse_dates=['date'])
```

```python
nvidia_controversies = pd.read_csv('nvidia_controversies_2023_2025_oct_drop_duplicates.csv', parse_dates=['date'])
nvidia_deals = pd.read_csv('nvidia_deals_2023_2025_oct_drop_duplicates.csv', parse_dates=['date'])
nvidia_financial = pd.read_csv('nvidia_financial_2023_2025_oct_drop_duplicates.csv', parse_dates=['date'])
nvidia_products = pd.read_csv('nvidia_products_2023jan_2025oct_drop_duplicates.csv', parse_dates=['date'])
nvidia_regulatory = pd.read_csv('nvidia_regulatory_2023_to_2025_oct_drop_duplicates.csv', parse_dates=['date'])
nvidia_competitors = pd.read_csv('nvidia_competitor_success_2023_2025_oct_drop_duplicates.csv', parse_dates=['date'])
```

Добавим столбец с названием компании и типом новости

```python
name = 'Nvidia'

nvidia_controversies[['company', 'type']] = [name, 'controversies']
nvidia_deals[['company', 'type']] = [name, 'deals']
nvidia_financial[['company', 'type']] = [name, 'financial']
nvidia_products[['company', 'type']] = [name, 'products']
nvidia_regulatory[['company', 'type']] = [name, 'regulatory']
nvidia_competitors[['company', 'type']] = [name, 'competitors']
```

```python
nvidia_data = pd.concat([nvidia_controversies, nvidia_deals, nvidia_financial, nvidia_products, nvidia_regulatory, nvidia_competitors])
nvidia_data = nvidia_data.sort_values('date')
```

```python
merged_macro = pd.merge_asof(left = nvidia_data, right= nums, on='date', direction='forward')
```

```python
nvidia_prices = pd.read_csv('nvda_stock_prices.csv', parse_dates=['date'])
```

```python
nvidia_prices_time = pd.merge_asof(left=merged_macro, right=nvidia_prices, on='date', direction='forward')
```
